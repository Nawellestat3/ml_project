When we do a linear regression we assume that the relationship between X (feature) and Y (target) is :

\begin{equation}
Y = \beta X + \epsilon
\end{equation}

Where $\epsilon$ is a noise, that usally follows a Normal distribution, and $\beta$ is the correlation coefficient.

```python
# let's simulate X and Y to illustrate the purpose of this model

import numpy as np
import matplotlib.pyplot as plt

n = 500
X = np.random.exponential(scale=2, size=n)
X = np.random.rand(n)

# relation between X and Y

def f(x):
    return 5*x

epsilon = np.random.normal(0, 1, n)

Y = f(X) + epsilon
```

```python
# let's visualise the shape of our random variables

plt.hist(X, color='teal')
plt.title("Distribution of X")
plt.show()
plt.hist(Y, color = 'gold')
plt.title(r"Distribution of $Y = X^2 + \epsilon$")
```

```python
# let's visualise the relationship between X and Y

plt.plot(X, Y, "o")
plt.title("Relationship between X and Y")
```
